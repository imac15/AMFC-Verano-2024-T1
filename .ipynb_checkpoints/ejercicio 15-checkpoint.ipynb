```python
# Arreglo a usar
datos = ['Ramirez','Garcia','Gonzales','Rodriguez','Perez','Martinez','Rodriguez','Martinez','Rodriguez','Hernandez','Garcia',
        'Garcia','Hernandez','Lopez','Perez','Gonzales','Martinez','Ramirez','Hernandez','Gonzales','Rodriguez','Sanchez','Hernandez',
        'Gonzales','Rodriguez','Martinez','Ramirez','Perez','Sanchez','Hernandez','Rodriguez','Lopez','Sanchez','Rodriguez','Ramirez',
        'Lopez','Gonzales','Lopez','Martinez','Lopez','Garcia','Sanchez','Gonzales','Gonzales','Lopez','Martinez','Gonzales','Martinez',
        'Ramirez','Gutierrez','Sanchez','Flores','Garcia','Martinez','Hernandez','Martinez','Garcia','Hernandez','Flores','Perez','Garcia',
        'Lopez','Gutierrez','Lopez','Gonzales','Hernandez','Garcia','Gutierrez','Garcia','Rodriguez','Hernandez','Sanchez','Gutierrez','Martinez',
        'Gonzales','Rodriguez','Hernandez','Gonzales','Garcia','Hernandez','Martinez','Garcia','Garcia','Hernandez','Garcia','Hernandez','Flores','Perez',
        'Hernandez','Ramirez','Lopez','Sanchez','Garcia','Flores','Perez','Lopez','Gonzales','Hernandez','Sanchez','Martinez']

datosF = p.format_list(datos) #Cambia el formato de los datos del arreglo

clases, fa_absoluta = r.frec_abs(datosF) #Calcula las clases unicas asi como su F_absoluta

clases_sorted = o.clases_str_sort(clases) #Ordena las clases por ordes alfabetico

fa_sorted = m.sort_clases_fa(clases, clases_sorted, fa_absoluta) #Ordena las F_absolutas

fr_relativa =  g.frecuencia_rel(fa_sorted) #Calculas las F_relativas

fr_acum = j.calcular_frecuencia_a(fr_relativa) #Calculas las F_acomuladas

tabla = ft.imptabla(clases_sorted, fa_sorted, fr_relativa, fr_acum) #Ordena los datos recabados en una tabla


print (tabla) #Impresion de la tabla

marcas_texto = clases_sorted #Da el valor de las clases ordenadas a marcas texto

# Definicion de 'titulos'
labelx = "Frecuencia"
labely = "Apellido"
titulo = "Apelidos"

# Impresion de las tablas correspondientes
histograma = ca.plot_hist(clases, fa_sorted, marcas_texto, labelx, labely, titulo)

pastel = re.plot_pie(fr_relativa,marcas_texto,titulo)
```
